Ans(1)

Pooling, specifically max pooling and average pooling, is a critical component in Convolutional Neural Networks (CNNs) that serves the purpose of down-sampling feature maps. Pooling offers several benefits in the context of CNNs:

Dimensionality Reduction:

Pooling reduces the spatial dimensions (width and height) of the feature maps, effectively reducing the number of parameters and computations in subsequent layers. This leads to faster training and inference, making the network more computationally efficient.

Translation Invariance:

Pooling helps to achieve translation invariance, meaning the network becomes less sensitive to small translations or shifts in the input. By selecting the maximum (max pooling) or average (average pooling) value in a local neighborhood, pooling retains the most dominant features, even if they slightly change position within the receptive field.

Noise Suppression:

Pooling can help reduce the impact of noise or minor variations in the input data. By considering the maximum or average value within a local region, pooling filters out small-scale variations or insignificant details, emphasizing the most important features.
Increased Receptive Field:

Pooling effectively increases the receptive field size. When a pooling layer follows a convolutional layer, the local regions from which pooling is performed correspond to larger regions in the original input. This allows the network to capture higher-level patterns and relationships among features.

Enhanced Computational Efficiency:

Smaller feature maps resulting from pooling require fewer computations in subsequent layers, making the network more computationally efficient. This is especially important in deep networks where computational resources can be a limiting factor.

Regularization and Generalization:

Pooling can serve as a form of regularization. By down-sampling the feature maps, it reduces the risk of overfitting by encouraging the network to focus on the most essential and generalizable features.

Hierarchical Feature Extraction:

Pooling is a key component in the hierarchical feature extraction process of CNNs. As the network progresses through layers, features become increasingly abstract and complex. Pooling helps in summarizing these features at different levels of abstraction.

Memory and Storage Efficiency:

Smaller feature maps generated by pooling require less memory and storage for storing intermediate results during training and inference, which can be advantageous for resource-constrained environments.

Max pooling and min pooling are two different types of pooling operations used in Convolutional Neural Networks (CNNs) for down-sampling feature maps. They serve similar purposes in terms of reducing the spatial dimensions of feature maps, but they differ in how they select and aggregate information from local regions. Here's the key difference between max pooling and min pooling:

Max Pooling:

In max pooling, for each local region within the feature map, the operation selects the maximum value from that region and passes it to the next layer.
Max pooling emphasizes the most dominant features within the region, making it particularly useful for recognizing patterns and details that stand out.
It is often preferred for its ability to capture the most important features, achieve translation invariance, and suppress noise and minor variations.
Max pooling is commonly used in CNN architectures for image-related tasks.

Min Pooling:

Min pooling is less common and somewhat unconventional in CNNs. It operates by selecting the minimum value from each local region within the feature map and passing it on.
Min pooling, in contrast to max pooling, focuses on emphasizing the least prominent features or the smallest values within a region.
This type of pooling is used less frequently and is not as popular as max pooling in practical CNN applications, primarily because it may not be as effective in recognizing important patterns and features.
Min pooling could have niche applications where emphasizing the smallest values within regions is relevant, but it's not typically employed in mainstream CNN architectures.

Padding is a crucial concept in Convolutional Neural Networks (CNNs) and plays a significant role in the network's architecture. Padding involves adding extra, zero-valued pixels around the edges of an input feature map before applying a convolution operation. It is essential for several reasons, and its significance lies in the following aspects:

Preserving Spatial Dimensions:

Padding ensures that the spatial dimensions of the feature maps remain the same after the convolution operation. This is particularly important when you want to maintain the same spatial resolution in the feature maps throughout the network. Without padding, the spatial dimensions would gradually shrink, leading to the loss of spatial information.

Centering the Convolutional Kernels:

When applying convolution, the kernel (or filter) is centered at each pixel location of the input feature map. Without padding, the kernel can only be centered on pixels that are sufficiently away from the edges. Padding allows the kernel to be centered at all pixels, ensuring that the convolution operation is applied uniformly across the entire feature map.

Mitigating Boundary Effects:

Convolution operations near the boundaries of feature maps can result in boundary effects, where only part of the kernel overlaps with the feature map. Padding helps mitigate these effects by ensuring that the entire kernel has coverage, leading to consistent feature extraction.

Controlling the Output Size:

Padding allows you to control the size of the output feature maps. By adjusting the amount of padding, you can determine the spatial dimensions of the output feature maps, which can be crucial in designing the architecture of a CNN.

Preserving Information at Multiple Scales:

In deep CNNs, multiple convolutional layers are often stacked. By using padding to maintain spatial dimensions, you enable the network to extract features at multiple scales, from fine details to larger structures. This is valuable for recognizing patterns and objects of varying sizes in the input.

Strides and Overlapping Receptive Fields:

Padding can be used in conjunction with varying strides to control the degree of overlap between receptive fields. This offers flexibility in terms of feature extraction and can be advantageous when adjusting the network's receptive field size.

Improving the Network's Capability:

Proper padding helps the CNN learn more discriminative features. With adequate padding, the network can focus on capturing information from the entire input region, including the edges and corners, which can be important for understanding the context and boundaries of objects.

Zero-padding:

Zero-padding involves adding extra, zero-valued pixels around the edges of the input feature map before applying a convolution operation.

The primary purpose of zero-padding is to control the spatial dimensions of the output feature map.

Zero-padding increases the size of the output feature map by adding an equal number of rows and columns around the edges. This increase in size is determined by the amount of padding added.

The formula to calculate the output size when using zero-padding is:
Output size = (Input size + 2 * Padding - Kernel size) / Stride + 1

Zero-padding is commonly used when you want to preserve the spatial dimensions of the input feature maps or when you need to control the output size.

Valid-padding:

Valid-padding (also known as 'no-padding') involves not adding any extra pixels around the input feature map. Instead, the convolutional kernel is centered only on valid positions within the input feature map.

The primary purpose of valid-padding is to reduce the spatial dimensions of the output feature map.

Valid-padding results in a smaller output feature map compared to the input size, as no additional pixels are added. The output size is determined by the convolutional kernel size and the stride
.
The formula to calculate the output size when using valid-padding is:
Output size = (Input size - Kernel size) / Stride + 1

Valid-padding is commonly used when you want to reduce the spatial dimensions of the feature maps or when you are not concerned about preserving the original spatial resolution.

It played a significant role in the development of deep learning and convolutional neural networks for image recognition tasks. Here's a brief overview of the LeNet-5 architecture:

Input Layer:

LeNet-5 takes grayscale images as input. The original LeNet-5 implementation used 32x32 pixel images, which were relatively small compared to modern CNNs.

Convolutional Layers:

LeNet-5 consists of two convolutional layers. The first convolutional layer uses a 5x5 kernel, while the second uses a 5x5 kernel followed by max-pooling.
These convolutional layers are responsible for feature extraction and are followed by activation functions, typically sigmoid or hyperbolic tangent (tanh) in the original design.

Pooling Layers:

After each convolutional layer, LeNet-5 includes subsampling (max-pooling) layers. These layers down-sample the spatial dimensions of the feature maps, helping to reduce computational complexity and enhance translation invariance.

Fully Connected Layers:

LeNet-5 includes three fully connected layers (also called dense layers). The first fully connected layer has 120 units, the second has 84 units, and the final layer has as many units as the number of classes in the classification task.
These fully connected layers serve as the network's decision-making components, and they are often followed by activation functions like the hyperbolic tangent (tanh).

Output Layer:

The output layer typically uses a softmax activation function for multi-class classification tasks. It provides probability scores for each class, allowing the network to make predictions.

Training:

LeNet-5 is typically trained using the backpropagation algorithm and is suitable for supervised learning tasks, particularly image classification.

Advantages of LeNet-5:

Pioneering Deep Learning: LeNet-5 was one of the earliest deep learning architectures that demonstrated the potential of deep neural networks for image classification tasks.

Effective Feature Extraction: LeNet-5's convolutional layers effectively extracted low-level features, such as edges and textures, followed by higher-level features, which allowed it to recognize patterns and objects.

Translation Invariance: The use of max-pooling layers contributed to translation invariance, making it robust to slight shifts in the position of objects within images.

Scalability: LeNet-5 provided a scalable architecture that could be adapted to various image classification tasks by adjusting the number of filters, units in fully connected layers, and output classes.

Inspiration for Modern CNNs: While it was relatively simple by modern standards, LeNet-5's principles and architectural design influenced the development of more advanced CNN architectures, including the popular AlexNet, VGGNet, and others.

Limitations of LeNet-5:

Small Input Size: LeNet-5 was designed for small 32x32 pixel grayscale images. It may not perform well on larger, high-resolution images common in modern computer vision tasks.

Sigmoid Activation: The use of sigmoid activation functions in LeNet-5 can suffer from the vanishing gradient problem and lead to slow convergence. Modern CNNs typically use rectified linear units (ReLU) for faster training.

Limited Capacity: The network's capacity may not be sufficient for complex image classification tasks with a large number of classes or highly detailed images.

Lack of Depth: LeNet-5 is relatively shallow by modern standards. Deep CNNs have demonstrated better performance in capturing hierarchical features and are more suited for complex recognition tasks.

Inefficient Use of Parallelization: The design of LeNet-5 is not well-suited for modern hardware acceleration, such as GPUs and TPUs, which are commonly used in training deep networks.

Overfitting: In certain cases, LeNet-5 may suffer from overfitting, particularly when dealing with small datasets or when the network architecture lacks appropriate regularization techniques.

```python
# importing the packages
import keras
from keras.datasets import mnist
from keras.layers import Conv2D,MaxPooling2D
from keras.layers import Dense,Flatten
from keras.models import Sequential
```

```python
# Loading MNIST dataset
(X_train,Y_train),(X_test,Y_test)=mnist.load_data()
```

```python
# Reshaping Training variables
X_train=X_train.reshape(X_train.shape[0],28,28,1)
X_test=X_test.reshape(X_test.shape[0],28,28,1)
```

```python
# one hot encoding
Y_train=keras.utils.to_categorical(Y_train,10)
Y_test=keras.utils.to_categorical(Y_test,10)
```

```python
# Normalization of training and test data
X_train=X_train/255
Y_test=Y_test/255
```

```python
# Building model
model=Sequential()
```

```python
# Adding convolutional layers to model
model.add(Conv2D(6,kernel_size=(5,5),activation='relu',input_shape=(28,28,1)))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Conv2D(16,kernel_size=(5,5),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
```

```python
# Adding ANN layers
model.add(Flatten())
model.add(Dense(120,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
# Displaying Summary
model.summary()
```

```python
#Compiling Model
model.compile(loss=keras.metrics.categorical_crossentropy,optimizer=keras.optimizers.Adam(),metrics=['accuracy'])
```

```python
# Training model
hist=model.fit(X_train,Y_train,batch_size=128,epochs=5,verbose=1,validation_data=(X_test,Y_test))
```

```python
# creating dataframe
import pandas as pd
df=pd.DataFrame(hist.history)
df
```

```python
# Drawing plot for dataframe
df.plot()
```

AlexNet is a deep convolutional neural network (CNN) architecture that gained widespread attention and significantly advanced the field of computer vision. It was developed by Alex Krizhevsky, Ilya Sutskever, and Geoffrey Hinton in 2012. AlexNet won the ImageNet Large Scale Visual Recognition Challenge in 2012, demonstrating a significant breakthrough in image classification tasks. Here's an overview of the AlexNet architecture:

1.Input Layer:

AlexNet accepts RGB images as input. It takes 224x224-pixel images, which is a larger input size compared to previous models at the time.

2.Convolutional Layers:

AlexNet consists of five convolutional layers, which are responsible for feature extraction. These convolutional layers are followed by ReLU (Rectified Linear Unit) activation functions to introduce non-linearity.
The first convolutional layer uses a large 11x11 filter with a stride of 4, while the subsequent layers use smaller 3x3 filters.
Local Response Normalization (LRN) was applied after the first and second convolutional layers to encourage competition between features and improve generalization.

3.Max-Pooling Layers:

After the first and second convolutional layers, max-pooling layers are applied to down-sample the feature maps and reduce spatial dimensions.
Max-pooling helps make the network translation invariant and computationally efficient.

4.Fully Connected Layers:

AlexNet includes three fully connected layers (also known as dense layers), which serve as the decision-making components of the network.
The first two fully connected layers consist of 4096 units each, followed by ReLU activation functions.
The final fully connected layer has as many units as the number of classes in the classification task (e.g., 1000 for ImageNet) and is followed by a softmax activation function.

5.Dropout:

Dropout was introduced in AlexNet as a regularization technique to reduce overfitting. It randomly drops a fraction of neurons during training, preventing the network from becoming overly reliant on specific features.

6.Data Augmentation:

AlexNet used data augmentation, which involves applying random transformations to the training data, such as cropping and horizontal flipping. This technique helps increase the diversity of the training dataset and improves generalization.

7.Training:

The network is trained using stochastic gradient descent (SGD) with a relatively high learning rate, which was reduced during training to fine-tune the model.
AlexNet's performance was significantly boosted by the use of graphical processing units (GPUs) for training, which was not as common at the time.

8.Output Layer:

The output layer uses a softmax activation function to produce class probabilities, allowing the network to make predictions for image classification.

AlexNet achieved breakthrough performance in image classification primarily due to several architectural innovations that set it apart from previous convolutional neural networks (CNNs). These innovations played a pivotal role in improving the model's accuracy and robustness. Here are the key architectural innovations introduced in AlexNet:

Deep Architecture:

AlexNet was one of the first deep CNNs. It utilized a deep architecture with multiple convolutional and fully connected layers, effectively capturing hierarchical features. Prior models tended to be shallower.

Large Convolutional Kernels:

ReLu Activation Function:

AlexNet used the Rectified Linear Unit (ReLU) activation function instead of traditional sigmoid or hyperbolic tangent (tanh) functions. ReLU introduces non-linearity and helps mitigate the vanishing gradient problem, enabling faster convergence during training.

Local Response Normalization (LRN):

LRN was applied after the first and second convolutional layers. LRN encouraged competition between feature maps and helped improve the network's generalization and overall performance.

Dropout Regularization:

AlexNet introduced dropout, a regularization technique that randomly deactivates a fraction of neurons during training. This helps prevent overfitting and encourages the network to learn more robust and generalizable features.

Data Augmentation:

Data augmentation techniques, such as random cropping and horizontal flipping, were employed during training. This augmented dataset diversity and improved the model's ability to generalize to variations in the input data.

Parallel Processing with GPUs:

AlexNet's training significantly benefited from the use of graphical processing units (GPUs), which accelerated the computation of forward and backward passes. This allowed the model to be trained on a larger dataset and more extensive architecture.

Large-Scale Training on ImageNet:

AlexNet was trained on the ImageNet Large Scale Visual Recognition Challenge dataset, which contained over one million images across thousands of classes. The scale of the dataset contributed to the network's ability to recognize a wide range of objects and patterns.

Ensemble of Models:

During the ImageNet challenge, AlexNet used an ensemble of multiple models and averaged their predictions. This ensemble approach further improved the model's accuracy.

Efficient Use of CNN Architectures:

AlexNet's efficient use of CNN architectures paved the way for future developments in CNN design. The convolutional and pooling layers were organized in a way that balanced feature extraction and dimension reduction.

In the AlexNet architecture, convolutional layers, pooling layers, and fully connected layers play distinct and complementary roles in the network's overall functionality. These layers work together to extract features from input images and make predictions. Here's a discussion of their roles in AlexNet:

Convolutional Layers:

Feature Extraction: The primary role of convolutional layers is feature extraction. These layers apply convolution operations to the input images, which involve sliding learnable filters (kernels) over the input data to detect local patterns, edges, textures, and higher-level features. In AlexNet, the use of deep convolutional layers helped capture hierarchical features, ranging from simple edges to complex object parts.

Non-linearity: After each convolution operation, AlexNet applies a Rectified Linear Unit (ReLU) activation function. ReLU introduces non-linearity, making the network capable of learning complex, non-linear relationships in the data. This helps in faster convergence during training.

Local Feature Hierarchies: Convolutional layers help in creating local feature hierarchies by extracting information from small local receptive fields. Deeper convolutional layers capture increasingly abstract and complex features.
Weight Sharing: Convolutional layers in AlexNet employ weight sharing, meaning the same set of filters is applied across the entire input image. This reduces the number of learnable parameters and encourages the network to recognize similar patterns in different regions of the input.

Pooling Layers:

Spatial Down-Sampling: Pooling layers are used for spatial down-sampling. They reduce the spatial dimensions of the feature maps produced by the convolutional layers. In AlexNet, max-pooling is applied, which selects the maximum value within a local region (receptive field) and retains it as the down-sampled representation.

Translation Invariance: Max-pooling contributes to translation invariance, allowing the network to recognize patterns and features regardless of their exact position within the input. This is particularly useful in object recognition tasks.

Computation Reduction: Pooling layers also reduce the computational complexity of the network by creating smaller feature maps, which require fewer parameters and computations in subsequent layers.
Feature Retention: Despite down-sampling, pooling layers retain the most salient features, ensuring that the network still focuses on critical information.

Fully Connected Layers:

Decision Making: Fully connected layers, also known as dense layers, serve as the decision-making components of the network. They take high-level features extracted by the convolutional and pooling layers and make class predictions.

Generalization: These layers capture complex relationships between features and learn to make predictions for various classes. The final fully connected layer has as many units as there are classes in the classification task, and it provides class probability scores.

Activation Functions: In AlexNet, fully connected layers are followed by activation functions (typically ReLU) to introduce non-linearity and make the network capable of modeling complex decision boundaries.


```python
# importing libaraies
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import os
import time
```

```python
# loading dataset
(train_images, train_labels), (test_images, test_labels) = keras.datasets.cifar10.load_data()
```

```python
# defining classes
CLASS_NAMES= ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
```

```python
# data processong
train_ds=tf.data.Dataset.from_tensor_slices((train_images,train_labels))
test_ds=tf.data.Dataset.from_tensor_slices((test_images,test_labels))
```

```python
# showing dataset
plt.figure(figsize=(30,30))
for i,(image,label) in enumerate(train_ds.take(20)):
    #print(label)
    ax=plt.subplot(5,5,i+1)
    plt.imshow(image)
    plt.title(CLASS_NAMES[label.numpy()[0]])
    plt.axis('off')
```

```python
# getting image label
def process_image(image,label):
    image=tf.image.per_image_standardization(image)
    image=tf.image.resize(image,(64,64))

    return image,label
```

```python
# checking train and test size
train_ds_size=tf.data.experimental.cardinality(train_ds).numpy()
test_ds_size=tf.data.experimental.cardinality(test_ds).numpy()
print('Train size:',train_ds_size)
print('Test size:',test_ds_size)
```

```python
# applying preprocessing batch and suffluning
train_ds=(train_ds
          .map(process_image)
          .shuffle(buffer_size=train_ds_size)
          .batch(batch_size=32,drop_remainder=True)
         )
test_ds=(test_ds
          .map(process_image)
          .shuffle(buffer_size=test_ds_size)
          .batch(batch_size=32,drop_remainder=True)
         )
```

```python
# Creating model
model=keras.models.Sequential([
    keras.layers.Conv2D(filters=128, kernel_size=(11,11), strides=(4,4), activation='relu', input_shape=(64,64,3)),
    keras.layers.BatchNormalization(),
    keras.layers.MaxPool2D(pool_size=(2,2)),
    keras.layers.Conv2D(filters=256, kernel_size=(5,5), strides=(1,1), activation='relu', padding="same"),
    keras.layers.BatchNormalization(),
    keras.layers.MaxPool2D(pool_size=(3,3)),
    keras.layers.Conv2D(filters=256, kernel_size=(3,3), strides=(1,1), activation='relu', padding="same"),
    keras.layers.BatchNormalization(),
    keras.layers.Conv2D(filters=256, kernel_size=(1,1), strides=(1,1), activation='relu', padding="same"),
    keras.layers.BatchNormalization(),
    keras.layers.Conv2D(filters=256, kernel_size=(1,1), strides=(1,1), activation='relu', padding="same"),
    keras.layers.BatchNormalization(),
    keras.layers.MaxPool2D(pool_size=(2,2)),
    keras.layers.Flatten(),
    keras.layers.Dense(1024,activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1024,activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(10,activation='softmax')
])
```

```python
# Compiling model
model.compile(
    loss='sparse_categorical_crossentropy',
    optimizer=tf.optimizers.SGD(lr=0.001),
    metrics=['accuracy']
)
```

```python
# checking model summary
model.summary()
```

```python
# Training model
history=model.fit(
    train_ds,
    epochs=10,
    validation_data=test_ds,
    validation_freq=1
)
```

```python
# visualising result of training model
f,ax=plt.subplots(2,1,figsize=(10,10))

#Assigning the first subplot to graph training loss and validation loss
ax[0].plot(model.history.history['loss'],color='b',label='Training Loss')
ax[0].plot(model.history.history['val_loss'],color='r',label='Validation Loss')

#Plotting the training accuracy and validation accuracy
ax[1].plot(model.history.history['accuracy'],color='b',label='Training  Accuracy')
ax[1].plot(model.history.history['val_accuracy'],color='r',label='Validation Accuracy')

plt.legend()
```

```python
# printing accuracy score
print('Accuracy Score = ',np.max(history.history['val_accuracy']))
```
